## PCA Clustring for BMW sales dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## import BMW sales dataset

```python
df= pd.read_csv('BMW sales data (2010-2024).csv')
df.head()
```

```python
df['Sales_Classification'].value_counts()
```

## Normalizing data

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = ["Model", "Region", "Color", "Fuel_Type", "Transmission"]
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
x= df.drop('Sales_Classification',axis=1)
y= df['Sales_Classification'].map({"Low": 0, "High": 1})
y = LabelEncoder().fit_transform(y)
```

## Split train\test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

## Prediction with ANN

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=100, max_iter=150)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

## Prediction with PCA

```python
from sklearn.decomposition import PCA
pca= PCA(n_components=4)
pca.fit(x_train)

nx_train=pca.transform(x_train)
nx_test=pca.transform(x_test)
```

## Calculating Accuracy, Precision, Recall, Confusion Matrix ANN

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("\nConfusion Matrix:\n", cm)
```

## Calculating Accuracy, Precision, Recall, Confusion Matrix PCA

```python
model = MLPClassifier(hidden_layer_sizes=100, max_iter=150)
model.fit(nx_train, y_train)
ny_pred = model.predict(nx_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
accuracy = accuracy_score(y_test, ny_pred)
precision = precision_score(y_test, ny_pred)
recall = recall_score(y_test, ny_pred)
cm = confusion_matrix(y_test, ny_pred)
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("\nConfusion Matrix:\n", cm)
```
